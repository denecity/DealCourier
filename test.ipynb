```python
import numpy as np
import matplotlib.pyplot as plt


total = 100000
counts = 20
ratios = []

for count in range(counts):
    counter = 0
    print(np.arange(count+1))
    print(np.random.permutation(np.arange(count+1)))


    for j in range(total):
        ls = np.arange(count+1)
        ls_perm = np.random.permutation(ls)

        for i in range(len(ls)):
            if ls[i] == ls_perm[i]:
                print()
                counter += 1
                break

    ratios.append(counter/total)
    print(counter/total)


plt.plot(ratios)
```
